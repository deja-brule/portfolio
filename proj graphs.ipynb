```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import Series, DataFrame
import numpy as np

SnowDF = pd.read_excel('snow.xlsx',
                       skiprows=(0,1), header=None,
                       names=['Month','Inches','Month2','Inches2'],
                       usecols='A:D')
print(SnowDF)
print()
x1=[1,2,3,4,5,6,7,8]
y1=[15.2,11.4,34.7,35.1,76.0,6,10,0]
x2=[1,2,3,4,5,6,7,8]
y2=[3,17,81,77.5,90,16,0,0]
xlabel=('Oct','Nov','Dec','Jan','Feb','Mar','Apr','May')

plt.plot(x1, y1, label = '2020-2021 Season', color='teal')
plt.plot(x2, y2, label = '2021-2022 Season', color='turquoise')
plt.xlabel('Month', color='steelblue',fontsize='15',horizontalalignment='center',verticalalignment='top')
plt.ylabel('Inches of Snow', color='steelblue', fontsize='15',verticalalignment='bottom')
plt.title('Snowfall in Houghton, Michigan', color='steelblue', fontweight='bold', fontsize='20')

plt.legend()
plt.show()
print()
print()

HockeyDF = pd.read_excel('hockey.xlsx',
                         skiprows=(0,1), header=None,
                         names=['Game','Points','Result'],
                         usecols='A:C')
x = [HockeyDF['Result']]
y = [HockeyDF['Points']]

plt.scatter(x, y, color='gold', marker='d')
plt.title("DU Hockey Team Points Scored v. Game Result", fontsize='20', fontweight='bold', color='maroon')
plt.ylabel("Points Scored", size='15', color='maroon')
plt.xlabel("Result (0 = Loss, 1 = Win)", size='15', color='maroon')
plt.show()
```
